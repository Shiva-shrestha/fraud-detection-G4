## Single data prediciton

```python
def single_predict(input_data):
    # Convert input_data to a DataFrame with a single row
    df1 = pd.DataFrame([input_data])

    feature_columns = model.named_steps['preprocessor'].transformers_[0][2] + \
                      model.named_steps['preprocessor'].transformers_[1][2]
    
    X_test_preprocessed = df1[feature_columns]
    
    # Predict using the trained model
    prediction = model.predict(X_test_preprocessed)

    return prediction[0]
```
