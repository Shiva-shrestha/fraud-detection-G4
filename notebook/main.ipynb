```python
import glob
import sys

# Add the path to your folder
sys.path.append("../core/")

import preprocessing
import model_training
import inference
```

```python
def train_model():
    # Define the path to the training datasets
    train_files = glob.glob('../data/splited_data/*.csv')

    # Preprocess the data
    X, y = preprocessing.preprocess_data(train_files)

    # Train the model and save it
    model_path = model_training.train_and_save_model(X, y, model_filename='../model/model.joblib')
    
    print(f"Model saved to {model_path}")

def test_model():
    # Define the path to the test dataset
    test_file_path = '../data/fraudTest.csv'

    # Predict on new data using the saved model
    fraud_predictions = inference.predict_on_new_data(model_filename='../model/model.joblib', test_file_path=test_file_path)

    # Display fraudulent predictions
    print(fraud_predictions.head())

    return fraud_predictions
```

```python
# Train the model
train_model()
```
