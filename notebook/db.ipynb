```python
import psycopg2
import pandas as pd

input_data = pd.DataFrame({
    'merchant': ['Merchant A'],
    'category': ['Category 1'],
    'amt': [100.50],
    'gender': ['M'],
    'lat': [40.7128],
    'long': [-74.0060],
    'city_pop': [8000000],
    'job': ['Engineer'],
    'unix_time': [1625097600],
    'merch_lat': [40.730610],
    'merch_long': [-73.935242],
    'trans_date_trans_time': ['2021-01-01 00:00:00'],
    'is_fraud': [0]
})

# Define connection parameters
conn_params = {
    "host": "localhost",  # Database host
    "port": "5432",       # Port number, default is 5432
    "database": "Fraud detection",  # Database name
    "user": "postgres",           # Username
    "password": "1234"        # Password
}


# Establish the connection
try:
    # Connect to PostgreSQL
    connection = psycopg2.connect(**conn_params)
    cursor = connection.cursor()
    
    # SQL command to create the table 
    insert_query = '''
    INSERT INTO predict_table (merchant, category, amt, gender, lat, long, city_pop, job, unix_time, merch_lat, merch_long, trans_date_trans_time,is_fraud)
    VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s,%s);
    '''
    
    # Iterate over the DataFrame rows and insert each row into the database
    for index, row in input_data.iterrows():
        cursor.execute(insert_query, (
            row['merchant'], row['category'], row['amt'], row['gender'], 
            row['lat'], row['long'], row['city_pop'], row['job'], 
            row['unix_time'], row['merch_lat'], row['merch_long'], row ['trans_date_trans_time'],row['is_fraud']
        ))
    
    # Commit the transaction
    connection.commit()
    print("DataFrame inserted successfully into simple_transactions table")

    select_query = ''' SELECT * FROM predict_table WHERE is_fraud=1 LIMIT 10'''
    cursor.execute(select_query)
    data_ = cursor.fetchall()

    print('Records:\n')
    for row in data_:
        print(row)
    
    
except Exception as error:
    print("Error ", error)

finally:
    # Close the connection
    if connection:
        cursor.close()
        connection.close()
        print("PostgreSQL connection is closed")
```
